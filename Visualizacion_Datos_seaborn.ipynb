```python
import pandas as pd
```

```python
datos= pd.read_csv('credit_card.csv')
datos.head()
```

```python
datos.info()
```

```python
datos.columns
```

```python
dic_columnas = {
    'LIMIT_BAL': 'limite',
    'CHECKING_ACCOUNT': 'cuenta_corriente',
    'EDUCATION': 'escolaridad',
    'MARRIAGE': 'estado_civil',
    'AGE': 'edad',
    'BILL_AMT': 'valor_factura',
    'PAY_AMT': 'valor_pago',
    'DEFAULT': 'moroso'
}
```

```python
tarjetas = datos.rename(columns= dic_columnas)
```

```python
tarjetas.head()
```

```python
tarjetas.cuenta_corriente.unique()
```

```python
dic_cuentas= {
    'Yes':'Si',
    'No': 'No'
}
```

```python
tarjetas.cuenta_corriente.map(dic_cuentas)
```

```python
tarjetas.cuenta_corriente= tarjetas.cuenta_corriente.map(dic_cuentas)
```

```python
tarjetas.head()
```

```python
tarjetas.escolaridad.unique()
```

```python
dic_escolaridad = {
    '2.University':'2.Universidad',
    '3.Graduate School':'Pos_Graduacion',
    '1.High School': 'Colegio'
}
```

```python
tarjetas.escolaridad= tarjetas.escolaridad.map(dic_escolaridad)
```

```python
tarjetas.head()
```

```python
tarjetas.estado_civil.unique()
```

```python
dic_estado_civil= {
    'Married':'Casado',
    'Single': 'Soltero'
}
```

```python
tarjetas.estado_civil= tarjetas.estado_civil.map(dic_estado_civil)
```

```python
tarjetas.head()
```

```python
#importamos seaborn
import seaborn as sns
```

```python
#graficar una variable y comparamos
sns.displot(data=tarjetas, x='limite',  hue='escolaridad')
```

```python
#creando columna con variable numerica
tarjetas['iu']= tarjetas['valor_factura']/ tarjetas['limite']
```

```python
tarjetas.head()
```

```python
sns.displot(data=tarjetas, x='iu')
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

la distribucion de los grupos morosos y no morosos es la misma

H<sup>null<sup>

la distribucion de los grupos morosos y no morosos no es la misma

```python
moroso= tarjetas.query('moroso==1').valor_factura
```

```python
moroso
```

```python
no_moroso= tarjetas.query('moroso==0').valor_factura
```

```python
no_moroso
```

```python
resultado=ranksums(moroso, no_moroso)
resultado
```

```python
print('El p_value de nuestro test es {}'.format(resultado.pvalue)) #acepta la hipotesis alternativa
```
